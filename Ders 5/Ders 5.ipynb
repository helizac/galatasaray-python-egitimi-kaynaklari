# Logarithmic Time — O(log n)

```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0
    
    while low <= high:
        mid = (high + low) // 2
    # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1
    # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1
```

# Quadratic Time — O(n²)

```python
def bubble_sort(data, value):
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(data)-1):
            if data[i] > data[i+1]:
                data[i], data[i+1] = data[i+1], data[i]
                swapped = True
```

# Exponential Time — O(2^n)

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

# Factorial — O(n!)

2! = 2 x 1 = 2
3! = 3 x 2 x 1 = 6
4! = 4 x 3 x 2 x 1 = 24
5! = 5 x 4 x 3 x 2 x 1 = 120
6! = 6 x 5 x 4 x 3 x 2 x 1 = 720
7! = 7 x 6 x 5 x 4 x 3 x 2 x 1 = 5.040
8! = 8 x 7 x 6 x 5 x 4 x 3 x 2 x 1 = 40.320

```python
def heap_permutation(data, n):
    if n == 1:
        return
    
    for i in range(n):
        heap_permutation(data, n - 1)
        if n % 2 == 0:
            data[i], data[n-1] = data[n-1], data[i]
        else:
            data[0], data[n-1] = data[n-1], data[0]
```

```python
data = [1, 2, 9, 8, 3, 4, 7, 6, 5]
value = random.choice(data)

for function in [get_first, binary_search, linear_search, merge_sort, bubble_sort, fibonacci, heap_permutation]:
    t0 = time.time()
    if function == fibonacci:
        fibonacci(len(data))
    else:
        function(data, value)
    t1 = time.time()
    print(str(function) + ": " + str(t1-t0))
```
