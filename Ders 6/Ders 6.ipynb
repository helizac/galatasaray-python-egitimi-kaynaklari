# Linear Search

```python
def linear_search(L, e):
    found = False
    for i in range(len(L)):
        if e == L[i]:
            found = True
            return i
```

# Binary Search

```python
liste = [1,3, 5, 10,25,2, 8]
linear_search(liste, 2)
```

```python
def binarySearch(array, x, low, high):

    # Repeat until the pointers low and high meet each other
    while low <= high:

        mid = low + (high - low)//2

        if array[mid] == x:
            return mid

        elif array[mid] < x:
            low = mid + 1

        else:
            high = mid - 1

    return -1
```

# Bubble Sort

```python
def bubble_sort(data):
    swapped = True
    while:
        swapped = False
        for i in range(len(data) - 1):
            if data[i] > data[i+1]:
                data[i], data[i+1] = data[i+1], data[i]
                swapped = True
```
